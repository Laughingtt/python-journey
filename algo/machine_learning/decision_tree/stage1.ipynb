```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import sklearn.datasets as datasets
```

```python
iris = datasets.load_iris()
print(iris.keys())
X = iris['data']
Y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
clf = DecisionTreeClassifier(criterion="gini",max_depth=4)
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
#决策树是什么样，绘制图形 
import graphviz
from sklearn import tree
dot_data = tree.export_graphviz(clf, out_file=None,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
graph = graphviz.Source(dot_data)
graph
```

```python




#随机森林
```

```python
from sklearn.ensemble import RandomForestClassifier
fore = RandomForestClassifier(n_estimators=100)
fore.fit(x_train,y_train)
fore.score(x_test,y_test)
```

```python
fore.n_estimators
```

```python
fore.estimators_
```

```python
forest =tree.export_graphviz(fore.estimators_[0],out_file=None,filled=True)
graphviz.Source(forest)
```

```python
#Adaboost  增强树，由多个弱项联合组成强项

from sklearn.ensemble import AdaBoostClassifier
```
